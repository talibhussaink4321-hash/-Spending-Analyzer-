```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('My first project.csv')
df_clean = df.dropna()
print("Data Loaded!")
print(df_clean)
```

```python
# Fix column names - remove exta spaces
df_clean.columns = df_clean.columns.str.strip()

# Now run analysis
print("SPENDING ANALYSIS")
print("===================")
print("Total Spent: ",df_clean['Money Spend'].sum())
print("Average/Day: ", round(df_clean['Money Spend'].mean(),2))
print("Highest Spend ", df_clean['Money Spend'].max())
print("Lowest Spend ", df_clean['Money Spend'].min())
    
```

```python
df_clean.plot(kind='bar',x='Day',
           y='Money Spend',
           color='steelblue',
           title='Daily Spending')
plt.tight_layout()
plt.show()
                                               
```

```python
# Pie Chart
plt.pie(df_clean['Money Spend'],
        labels=df_clean['Category'],
        autopct='%1.1f%%')
plt.title('Spending by Category')
plt.show()
```

```python
top = df_clean.loc[df_clean['Money Spend'].idxmax()]
low = df_clean.loc[df_clean['Money Spend'].idxmin()]

print("Highest Spending Day", top['Day'],"-",top['Money Spend'])
print("Lowest Spending Day", low['Day'],"-",low['Money Spend'])
print("Advice: Reduce spending on",top['Category'])
```
